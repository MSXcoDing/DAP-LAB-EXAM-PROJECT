```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Billionaires Statistics Dataset.csv')
```

```python
billionaires_by_country = data.groupby('country')['personName'].count()
country_with_most_billionaires = billionaires_by_country.idxmax()
number_of_billionaires = billionaires_by_country.max()
print(f"The country with the highest number of billionaires is {country_with_most_billionaires} with {number_of_billionaires} billionaires.")
```

```python
# Group the DataFrame by 'category' and count the number of billionaires in each category
billionaires_by_category = data.groupby('category')['personName'].count()

# Plot the distribution of billionaires across categories
plt.figure(figsize=(10, 6))
billionaires_by_category.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Distribution of Billionaires Across Categories')
plt.xlabel('Category')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
male_count = data[data['gender'] == 'M'].shape[0]
female_count = data[data['gender'] == 'F'].shape[0]

# Plot the comparison between male and female billionaires
labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['lightblue', 'pink']

plt.figure(figsize=(6, 6))
plt.bar(labels, sizes, color=colors)
plt.title('Comparison of Male and Female Billionaires')
plt.xlabel('Gender')
plt.ylabel('Number of Billionaires')
plt.show()
```

```python

current_year = 2024  # Assuming the current year is 2024
data['age'] = current_year - data['birthYear']

# Plot the age distribution of billionaires
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Billionaires')
plt.xlabel('Age')
plt.ylabel('Number of Billionaires')
plt.grid(True)
plt.show()
```
